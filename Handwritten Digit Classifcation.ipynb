## Handwritten Digit Classifcation MNIST dataset


The MNIST dataset is a widely used benchmark dataset in machine learning and computer vision. It consists of a collection of 60,000 grayscale images of handwritten digits (0-9) for training and 10,000 images for testing. Each image is a 28x28 pixel square, representing a single digit. The dataset is labeled, meaning each image is associated with the corresponding digit it represents. MNIST is often used as a starting point for learning and evaluating various image classification algorithms, and it has played a significant role in advancing the field of deep learning.

## Data Preprocessing

```python
# Train and test data scaling
xtrain_scale = xtrain/255
xtest_scale = xtest/255
```

```python
# conveting to_categorical ytrain and ytest it is like get_dummies that
from keras.utils import to_categorical
ytrain_cat = to_categorical(ytrain)
ytest_cat = to_categorical(ytest)
```

## Start training CNN Neural Network


1. Sequential: Sequential is a linear stack of layers used for building deep learning models in a sequential manner.

2. Convolution2D: A layer that applies convolutional operations to extract features from 2D input data.

3. MaxPooling2D: A layer that performs downsampling by selecting the maximum value within a specific window.

4. BatchNormalization: A layer that normalizes the input data to accelerate training and improve generalization.

5. Dropout: A layer that randomly deactivates a fraction of neurons during training to prevent overfitting.

6. Dense: A fully connected layer that applies linear operations to transform input data.

7. Flatten: A layer that reshapes multidimensional data into a 1D vector for further processing.

8. EarlyStopping: Stopping training when a monitored metric stops improving significantly

```python
from keras.models import Sequential
from keras.layers import Convolution2D,MaxPooling2D,BatchNormalization,Dropout,Dense,Flatten
from keras.callbacks import EarlyStopping
```

```python
xtrain.shape
```

```python
# initialize neural network with Sequential model
nn = Sequential()
```

```python
# Convolution Layer 1
nn.add(Convolution2D(filters=16,kernel_size=(3,3),input_shape=(28,28,1)))
nn.add(MaxPooling2D(pool_size=(2,2)))
nn.add(BatchNormalization())
nn.add(Dropout(0.2))

# Convolution Layer 2
nn.add(Convolution2D(filters=16,kernel_size=(3,3)))
nn.add(MaxPooling2D(pool_size=(2,2)))
nn.add(BatchNormalization())
nn.add(Dropout(0.2))

# Flatten Layer
nn.add(Flatten())

# Hidden Dense Layers
nn.add(Dense(128,activation="relu"))
nn.add(Dense(128,activation="relu"))


# Output Layer
nn.add(Dense(10,activation="softmax"))
```

```python
nn.summary()
```

```python
#Stopping training when a monitored metric stops improving significantly.
early_stop = EarlyStopping(monitor="val_loss",patience=3)
```

```python
#Configuring parameters and settings for optimization, loss, metrics, and other options in a deep learning model.
nn.compile(loss="categorical_crossentropy",optimizer="adam",
           metrics=["accuracy"])
```

```python
hist = nn.fit(xtrain_scale,ytrain_cat,
             validation_split=0.2,
             epochs=50,
             batch_size=32,
              callbacks=[early_stop])
```

## Visualazation learning curve

```python
plt.figure(figsize=(10,6))
plt.plot(hist.history["accuracy"])
plt.plot(hist.history["val_accuracy"])
plt.xlabel("Epoch")
plt.ylabel("accuracy")
plt.title("Accuracy_curve")
plt.show()
```

## Evaluate

```python
nn.evaluate(xtrain_scale,ytrain_cat)
```

```python
nn.evaluate(xtest_scale,ytest_cat)
```

## Create a function to predict images

```python
import keras
import cv2 
from keras.models import load_model
my_model = load_model("cnn_model.h5")

def predict_image(path,model):
    img = cv2.imread(path,cv2.IMREAD_GRAYSCALE)
    img = ~img
    plt.imshow(img,cmap="gray")
    img_scaled = img/255
    img_scaled = np.array([img_scaled])
    prob = model.predict(img_scaled)
    pred = np.argmax(prob,axis=-1)
    print(f" The predicted number : {pred[0]} with probility {prob.max()}")
    return pred[0],prob.max()
```
